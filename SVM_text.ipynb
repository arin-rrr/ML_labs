# Импортируем данные (тексты)

```python
from sklearn import datasets
```

```python
dataset = datasets.fetch_20newsgroups(subset='all', categories=['alt.atheism', 'sci.space'])
```

```python
# dataset.data
dataset.target
```

# Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
texts = vectorizer.fit_transform(dataset.data)
features_name = vectorizer.get_feature_names_out()
```

# Кросс-валидация и подбор гиперпараметров

```python
from sklearn.model_selection import KFold, GridSearchCV
import numpy as np
from sklearn.svm import SVC
```

```python
cv = KFold(n_splits=5, shuffle=True, random_state=241)

grid = {'C': np.power(10.0, np.arange(-5, 6))}
clf = SVC(kernel='linear', random_state=241)

gs = GridSearchCV(clf, grid, scoring='accuracy', cv=cv)
```

```python
gs.fit(texts, dataset.target)
```

```python
# определяем лучший параметр С
gs.best_params_
```

# Обучаем SVM при лучшем параметре

```python
clf_best = SVC(kernel='linear', C=1.0, random_state=241)
```

```python
clf_best.fit(texts, dataset.target)
```

```python
coef = clf_best.coef_.toarray()[0]
```

```python
sorted_indices = np.argsort(np.abs(coef))[::-1]  # Индексы от большего к меньшему
sorted_words = features_name[sorted_indices]  # Слова, соответствующие индексам
top_words = sorted([word.lower() for word in sorted_words[:10]])

top_words
```
